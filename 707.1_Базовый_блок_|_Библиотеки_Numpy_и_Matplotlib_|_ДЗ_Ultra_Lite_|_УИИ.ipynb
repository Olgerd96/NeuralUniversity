```python
import matplotlib.pyplot as plt
month = [random.randint(18,33) for x in range(30)]
plt.figure(figsize=(14, 5))
plt.plot(month, label="avg day temp")
plt.legend()
plt.grid()
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Temperature graph");
```

* создайте список с именем **programmingLanguages**, содержащий следующие значения:

        ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']

* создайте список с именем **popularity**, состоящий из 6 произвольных значений (можно задать вручную). Данный список будет содержать значение популярности соответсвующего языка программироваия
* с помощью метода `plt.bar()` постройте гистограмму популярности языков программирования, передав в метод `plt.bar()` значения списков с названиями языков и значениями их популярности;
* с помощью методов `plt.xlabel()` и `plt.ylabel()` задайте подписи осей графика: "Языки" и "Популярность" соответственно;

---

```python
import matplotlib.pyplot as plt

programmingLanguages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [random.randint(5,25) for x in range(6)]
plt.figure(figsize=(8, 5))
plt.bar(x=programmingLanguages, height=popularity)
plt.xlabel("Language")
plt.ylabel("Popularity")
plt.title("Popularity pogramming languages\n(October 2023)")
```
